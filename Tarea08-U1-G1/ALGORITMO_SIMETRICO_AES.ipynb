```python
import time
import random
from Crypto.Cipher import AES
from Crypto.Util.Padding import pad, unpad
```

500 PALABRAS

```python
start_file = time.time()

f = open('500w.txt', 'r')
mensajeOriginal = f.read().encode()

end_file = time.time()
print("Tiempo de lectura para el archivo (500 palabras): " + format(end_file-start_file))
print("\r")

start_encrypt = time.time() 

valor=random.randint(1000000000000000,1234567812345678)
numero=str(valor)
password = bytes(numero,'utf-8') 
print("La contrasenia es:" +format(password))


aes = AES.new(password, AES.MODE_ECB)
encrypted_text = aes.encrypt(pad(mensajeOriginal, 32))

end_encrypt = time.time()
print("Tiempo de encriptacion para el archivo (500 palabras): " + format(end_encrypt-start_encrypt))
print(encrypted_text)
print("\r")

start_uncrypt = time.time()

uncrypted = aes.decrypt(encrypted_text)
uncrypted_text = unpad(uncrypted,32) 

end_uncrypt = time.time()
print("Tiempo de des-encriptacion para el archivo (500 palabras): " + format(end_uncrypt-start_uncrypt))
print(uncrypted_text)
print("\r")
```
